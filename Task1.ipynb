```python
# To load iris dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("Libraries are loaded!")
```

```python

# To see 1st five rows of iris dataset
df = sns.load_dataset('iris')
print(df.head())
```

```python
# To see summary statistics(mean, std, min, max)
print(df.info())
print(df.describe())
```

```python
import matplotlib.pyplot as plt
# Scatter plot: between sepal length and width
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width')
plt.show()
```

```python
# Histogram: To see value distribution
plt.title('Petal Length Distribution')
plt.show()
```

```python
# To check outliers
sns.boxplot(x='species', y='petal_width', data=df)
plt.title('Petal Width Box Plot')
plt.show()
```
